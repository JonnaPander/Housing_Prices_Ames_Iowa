# Data Cleanup

#### Outliers

```python
#Identify outliers below 100k saleprice

below_100k = train[(train['saleprice'] < 100000)][['id', 'saleprice']]
below_100k.sort_values('saleprice', ascending=True).head(15)
```

```python
# Drop 15 records for sales below 100k.
# ISSUE: could not return index of sales < 100k to list
index_drop_low = [1628, 183, 1309, 1292, 1518, 1564, 1428, 408, 103, 1089, 1301, 410, 2010, 1374, 1952]

for i in index_drop_low:
    train = train.drop([i])
```

```python
#Identify outliers above 300k saleprice

above_300k = train[(train['saleprice'] > 500000)][['id', 'saleprice']]
above_300k.sort_values('saleprice', ascending=False).head(15)
```

```python
# Drop records for sales above 500k
index_drop_high = [1671, 1964, 1692, 1796, 151, 1164, 623, 138, 823, 1592, 81, 800, 738, 1754, 1317]

for i in index_drop_high:
    train = train.drop([i])
```

#### Add Columns

```python
# Convert 'year_built', 'year_remodeled' to age

#train
train['age'] = train['yr_sold'] - train['year_built']

#test
test['age'] = train['yr_sold'] - train['year_built']
```

```python
# Check values in columns

#train
train[['age','yr_sold', 'year_built']].head()

#test
test[['age','yr_sold', 'year_built']].head()
```

```python
# Interaction term,'age * overall_qual', for train
train['age * overall_qual'] = train['age'] * train['overall_qual']

# Interaction term for test
test['age * overall_qual'] = test['age'] * test['overall_qual']
```

```python
# Scatter plot

sns.pairplot(train, x_vars=['age * overall_qual'], y_vars=['saleprice'])
plt.figure(figsize=(25,25));
```

```python
# Interaction term, 'gr_liv * full_bath', for train
train['gr_liv * full_bath'] = train['gr_liv_area'] * train['full_bath']

# Interation term for tst
test['gr_liv * full_bath'] = test['gr_liv_area'] * test['full_bath']
```

```python
# Scatter plot

sns.pairplot(train, x_vars=['gr_liv * full_bath'], y_vars=['saleprice'])
plt.figure(figsize=(25,25));
```

#### Dummify

```python
# Shape before dummies
train.shape
test.shape
```

```python
# List of remaining columns of type object

train.select_dtypes(include=['object']).columns
```

```python
# Dummify object datatypes in train

dummy_objects = ['ms_zoning', 'street', 'lot_shape', 'land_contour', 'utilities', 'lot_config', 'land_slope',
       'neighborhood', 'condition_1', 'condition_2', 'bldg_type',
       'house_style', 'roof_style', 'roof_matl', 'exterior_1st',
       'exterior_2nd', 'mas_vnr_type', 'exter_qual', 'exter_cond',
       'foundation', 'bsmt_qual', 'bsmt_cond', 'bsmt_exposure',
       'bsmtfin_type_1', 'bsmtfin_type_2', 'heating', 'heating_qc',
       'central_air', 'electrical', 'kitchen_qual', 'functional',
       'garage_type', 'garage_finish', 'garage_qual', 'garage_cond',
       'paved_drive', 'sale_type']


for col in dummy_objects:
    train = pd.get_dummies(train, columns=[col], prefix=col) 
```

```python
# Dummify test df
    
for col in dummy_objects:
    test = pd.get_dummies(test, columns=[col], prefix=col) 
```
